```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import wave as wave
from scipy import signal
import struct
```

```python
N= [100,1000,10000]
ranval = -5+ (5+5)*np.random.rand(N[2])
counts, bins = np.histogram(ranval,10)
print(counts)
relativefreq = counts/float(N[2])*100
fig, ax1 = plt.subplots()
ax1.hist(ranval,align="left",bins=bins)
ax2 = ax1.twinx()
x = range(-5,5)
ax2.plot(x,relativefreq,"r")
ax2.grid(True)
ax2.axis('tight')
plt.show()

moyenne = np.mean(ranval)
ecarttype = np.std(ranval)
mu = moyenne
var = ecarttype**2
errquad = (moyenne-0)**2
print("moyenne = ",moyenne)
print("ecart type = ",ecarttype)
print("mu = ",mu)
print("var = ",var)
print("erreur quadratique = ",errquad)
```

```python
#4
N= 10000
def F(x):
    if(x<0):
        return 1-F(np.abs(x))
    else:
        return 0.5*(1+np.sqrt(1-np.exp(-(x)**2*np.sqrt(np.pi/8))))
val= [F(x) for x in np.linspace(-5,5,N)]

plt.plot(val)
plt.show()
#b)
u1 = np.random.rand(N)
def invF(x):
    if(x<0.5):
        return -np.sqrt(-np.log(-(2*x-1)**2+1)/np.sqrt(np.pi/8))
    else:
        return np.sqrt(-np.log(-(2*x-1)**2+1)/np.sqrt(np.pi/8))
    
x = [invF(u) for u in u1]

plt.hist(x,100)
plt.show()
```
